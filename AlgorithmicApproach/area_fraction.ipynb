```python
import imagej
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# initialize ImageJ
ij = imagej.init('sc.fiji:fiji:2.14.0') 
print(f"ImageJ version: {ij.getVersion()}")
```

```python
def area_fraction(arr): #Returns area fraction of a binary array
    lenght = arr.shape[0]
    width = arr.shape[1]

    counts = np.bincount(arr.ravel())

    return counts[255] / (lenght * width)
```
