```python
import imagej
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# initialize ImageJ
ij = imagej.init('sc.fiji:fiji:2.14.0') 
print(f"ImageJ version: {ij.getVersion()}")
```

```python
def area_fraction(arr): #Returns area fraction of a binary array

    return np.sum(arr > 0) / arr.size
```

```python
path = r"C:\Users\magfa\Documents\Prosjekt\Prosjektbilder\1m_x100 (11).tif"
```

```python
imp = ij.IJ.openImage(path)

np_arr = np.asarray(ij.py.from_java(imp))

print(area_fraction(np_arr))
```

```python
import pandas as pd
def area_fractions_from_folder(folder_path, csv):
    results = []
    folder = Path(folder_path)

    for file in folder.iterdir():
        imp = ij.IJ.openImage(str(file.resolve()))
        arr = np.asarray(ij.py.from_java(imp))
        imp.close()
        results.append({"Filename:" : {file.stem}, "Area fraction" : area_fraction(arr)})

    df = pd.DataFrame(results)
    df.to_csv(csv, index=False)
    avg = df["Area fraction"].mean()
    print("Average area fraction:", avg)
```
