Plotting Loss and IoU

```python
import ast
import matplotlib.pyplot as plt

dic = []
lst = []
with open("brightness_aug_3_log_2025_10_23_17_02_13.txt") as f:
    for line in f:
        dic = ast.literal_eval(line)
        lst.append(dic)
print(lst)
```

```python
train_loss = []
test_loss = []
train_iou = []
test_iou = []
epoch = []

for i in range(len(lst)):
    train_loss.append(lst[i]["train_loss"])
    test_loss.append(lst[i]["test_loss"])
    train_iou.append(lst[i]["train_IoU"])
    test_iou.append(lst[i]["test_IoU"])
    epoch.append(lst[i]["epoch"])
```

```python
plt.style.use("fast")

plt.figure(figsize=(9, 7))
plt.plot(epoch, train_loss, label = "Train Loss")
plt.plot(epoch, test_loss, label = "Test Loss")
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize = 14)
plt.tick_params(axis='both', labelsize=12)
plt.title("Evolution of Loss Function", fontsize =  18)
plt.legend(fontsize = 14)
plt.savefig("Loss.pdf", dpi = 1300)
```

```python
plt.style.use("fast")

plt.figure(figsize=(9, 7))
plt.plot(epoch, train_iou, label = "Train IoU")
plt.plot(epoch, test_iou, label = "Test IoU")
plt.xlabel("Epochs", fontsize = 14)
plt.ylabel("IoU", fontsize = 14)
plt.tick_params(axis='both', labelsize=12)
plt.title("Evolution of 'Intersection over Union'", fontsize = 18)
plt.legend(fontsize = 14)
plt.savefig("IoU.pdf", dpi = 1300)
```
