Plotting Loss and IoU

```python
import ast
import matplotlib.pyplot as plt

dic = []
lst = []
with open("brightness_aug_3_log_2025_10_23_17_02_13.txt") as f:
    for line in f:
        dic = ast.literal_eval(line)
        lst.append(dic)
print(lst)
```

```python
train_loss = []
test_loss = []
train_iou = []
test_iou = []
epoch = []

for i in range(len(lst)):
    train_loss.append(lst[i]["train_loss"])
    test_loss.append(lst[i]["test_loss"])
    train_iou.append(lst[i]["train_IoU"])
    test_iou.append(lst[i]["test_IoU"])
    epoch.append(lst[i]["epoch"])
```

```python
dic = []
lst = []
with open("ny_modell_26_11_2_log_2025_11_26_09_51_20.txt") as f:
    for line in f:
        dic = ast.literal_eval(line)
        lst.append(dic)
print(lst)
```

```python
train_loss_h = []
test_loss_h = []
train_iou_h = []
test_iou_h = []
epoch_h = []

for i in range(len(lst)):
    train_loss_h.append(lst[i]["train_loss"])
    test_loss_h.append(lst[i]["test_loss"])
    train_iou_h.append(lst[i]["train_IoU"])
    test_iou_h.append(lst[i]["test_IoU"])
    epoch_h.append(lst[i]["epoch"])
```

```python
fig, axarr = plt.subplots(2, 2, figsize = (14, 7))
flat = axarr.flatten()
flat[0].plot(epoch, train_loss, label = "Train Loss")
flat[0].plot(epoch, test_loss, label = "Test Loss")
flat[0].set_ylabel("Loss")
flat[0].set_title("Loss - Gaussian/Threshold")
flat[0].legend()

flat[2].plot(epoch_h, train_loss_h, label = "Train Loss")
flat[2].plot(epoch_h, test_loss_h, label = "Test Loss")
flat[2].set_xlabel("Epochs")
flat[2].set_ylabel("Loss")
flat[2].set_title("Loss - Hand Made")
flat[2].legend()

flat[1].plot(epoch, train_iou, label = "Train IoU")
flat[1].plot(epoch, test_iou, label = "Test IoU")
flat[1].set_ylabel("IoU")
flat[1].set_title("IoU - Gaussian/Threshold")
flat[1].legend()

flat[3].plot(epoch_h, train_iou_h, label = "Train IoU")
flat[3].plot(epoch_h, test_iou_h, label = "Test IoU")
flat[3].set_xlabel("Epochs")
flat[3].set_ylabel("IoU")
flat[3].set_title("IoU - Hand Made")
flat[3].legend()

fig.savefig("Loss_IoU.pdf", dpi = 1300)
```
