```python
import tifffile
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

Figure showing training results for "thresholding + gaussian blur" model

```python
raw0 = tifffile.imread(r'Gaussian_Images\gaussian_raw1.tif')
raw1 = tifffile.imread(r'Gaussian_Images\gaussian_raw2.tif')
raw2 = tifffile.imread(r'Gaussian_Images\gaussian_raw3.tif')


result0 = tifffile.imread(r'Gaussian_Images\gaussian_result1.tif')
result1 = tifffile.imread(r'Gaussian_Images\gaussian_result2.tif')
result2 = tifffile.imread(r'Gaussian_images\gaussian_result3.tif')


label0 = tifffile.imread(r'Gaussian_Images\gaussian_label1.tif')
label1 = tifffile.imread(r'Gaussian_Images\gaussian_label2.tif')
label2 = tifffile.imread(r'Gaussian_images\gaussian_label3.tif')
```

```python
result0 = (result0 / result0.max() > 0).astype(int)
result1 = (result1 / result1.max() > 0).astype(int)
result2 = (result2 / result2.max() > 0).astype(int)

label0 = (label0 / label0.max() > 0).astype(int)
label1 = (label1 / label1.max() > 0).astype(int)
label2 = (label2 / label2.max() > 0).astype(int)
```

```python
temp_label0 = label0 * 2
temp_label1 = label1 * 2
temp_label2 = label2 * 2

temp_result0 = result0 * 3
temp_result1 = result1 * 3
temp_result2 = result2 * 3

overlay0 = temp_label0 + temp_result0
overlay1 = temp_label1 + temp_result1
overlay2 = temp_label2 + temp_result2

colors = ["black", "black", "red", "blue", "black", "green"]

cmap = ListedColormap(colors=colors)
print(np.unique(label0))
print(np.unique(result0))

print(np.unique(overlay0))
```

```python
fig, axarr = plt.subplots(3, 4)

flat = axarr.flatten()

for i in range(12):
    flat[i].set_xticks([])
    flat[i].set_yticks([])

flat[0].imshow(raw0, cmap="Greys_r")
flat[0].set_title("Input Image")
flat[1].imshow(result0, cmap="Greys_r")
flat[1].set_title("Prediction")
flat[2].imshow(label0, cmap="Greys_r")
flat[2].set_title("Label")
flat[3].imshow(overlay0, cmap=cmap)
flat[3].set_title("Overlay")
flat[4].imshow(raw1, cmap="Greys_r")
flat[4].set_title("Input Image")
flat[5].imshow(result1, cmap="Greys_r")
flat[5].set_title("Prediction")
flat[6].imshow(label1, cmap="Greys_r")
flat[6].set_title("Label")
flat[7].imshow(overlay1, cmap=cmap)
flat[7].set_title("Overlay")
flat[8].imshow(raw2, cmap="Greys_r")
flat[8].set_title("Input Image")
flat[9].imshow(result2, cmap="Greys_r")
flat[9].set_title("Prediction")
flat[10].imshow(label2, cmap="Greys_r")
flat[10].set_title("Label")
flat[11].imshow(overlay2, cmap=cmap)
flat[11].set_title("Overlay")

fig.subplots_adjust(left=0,bottom=0, right=1, top=1, hspace=0, wspace=0.2)

#fig.set_figheight(4)
#fig.set_figwidth(4)

plt.savefig("Gaussian_Model.pdf", dpi = 1300)
```
